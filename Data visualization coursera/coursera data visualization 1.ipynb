Artist layer

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas#anti grain geometry
from matplotlib.figure import Figure
fig = Figure()
canvas = FigureCanvas(fig)
```

```python
import numpy as np
x = np.random.randn(10000)
```

```python
ax = fig.add_subplot(111)#create an axex artist
```

```python
ax.hist(x,100)
```

```python
ax.set_title('Normal graph')
fig.savefig('Matplotlib histogram')
```
